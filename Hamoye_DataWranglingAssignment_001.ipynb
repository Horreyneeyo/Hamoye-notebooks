### Part One: Load Data

```python
import numpy as np
import pandas as pd
```

```python
url_CellPhonesData ='https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv'
CellPhonesData = pd.read_csv(url_CellPhonesData, error_bad_lines=False)
```

```python
url_PopulationData = 'https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv'
PopulationData = pd.read_csv(url_PopulationData, error_bad_lines=False)
PopulationData.head()
```

```python
url_CountryCodesData = 'https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv'
CountryCodesData = pd.read_csv(url_CountryCodesData, error_bad_lines=False)
CountryCodesData.head()
```

### Part Two: Join Data

```python
newdata_df= CellPhonesData.merge(PopulationData, on=['Country', 'year'], how='inner')
newdata_df.head()
newdata_df.shape
```

```python
newdata_df.head()
```

```python
CountryCodesData.rename(columns={'country':'geo', 'name':'Country'}, inplace='True')
CountryCodesData.head()
```

```python
merged_data_df = newdata_df.merge(CountryCodesData, on=['Country'], how='inner')
merged_data_df.head()
```

```python
merged_data_df.drop(['g77_and_oecd_countries', 'income_3groups', 'income_groups',
       'is--country', 'iso3166_1_alpha2', 'iso3166_1_alpha3',
       'iso3166_1_numeric', 'iso3166_2', 'landlocked', 'latitude', 'longitude',
       'main_religion_2008', 'un_sdg_ldc', 'un_sdg_region', 'un_state',
       'unicef_region', 'unicode_region_subtag', 'world_4region',
       'world_6region'], axis=1, inplace=True) 
```

```python
merged_data_df.shape
```

### Part Three: Make Features

```python
merged_data_df.tail()
```

```python
i = 0
average_list = []
while i <= 11639:
    average_list.append(merged_data_df.at[i,'cell_phones_total']/merged_data_df.at[i, 'population'])
    i+=1
merged_data_df['cellphones_per_person']=average_list
```

```python
merged_data_df.iloc[11157]
```

```python
check1 = merged_data_df['Country'].str.contains('United States')
check2 = merged_data_df['year'] == 2017
assert merged_data_df.loc[check1 & check2]['cellphones_per_person'].item() == 1.2307692307692308
```

```python
merged_data_df['geo'] = merged_data_df['geo'].str.upper()
merged_data_df.head()
```
